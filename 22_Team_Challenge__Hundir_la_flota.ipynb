## Cmo funciona el juego?
Vamos a realizar una versin que tiene algunas particularidades respecto al juego original, de manera que sea ms sencillo el desarrollo. Veamos cmo funciona:

1. Hay dos jugadores: t y la mquina
2. Un **tablero de 10 x 10** posiciones donde irn los barcos
3. Lo primero que se hace es colocar los barcos. Para este juego **los barcos se colocan de manera aleatoria. Ahora bien, puedes empezar colocando los barcos en unas posiciones fijas, que no cambien con cada partida, y despus implementarlo aleatoriamente, ya que es ms complejo. Los barcos son:**
    * 4 barcos de 1 posicin de eslora
    * 3 barcos de 2 posiciones de eslora
    * 2 barcos de 3 posiciones de eslora
    * 1 barco de 4 posiciones de eslora

4. Tanto t, como la mquina tenis un tablero con barcos, y se trata de ir "disparando" y hundiendo los del adversario hasta que un jugador se queda sin barcos, y por tanto, pierde.
5. Funciona por turnos y empiezas t.
6. En cada turno disparas a una coordenada (X, Y) del tablero adversario. **Si aciertas, te vuelve a tocar**. En caso contrario, le toca a la mquina.
7. En los turnos de la mquina, si acerta tambin le vuelve a tocar. Dnde dispara la maquina? A un punto aleatorio en tu tablero.
8. Si se hunden todos los barcos de un jugador, el juego acaba y gana el otro.

En [esta pgina](http://es.battleship-game.org/) podrs probarlo online.

## Desarrollo
Tendrs que desarrollar lo siguiente:
1. Necesitars un conjunto de **constantes**, donde tengas inventariados los barcos del juego, dimensiones y dems variables que no vayan a cambiar que tendris definidas en archivo de **variables.py**

2. Tendrs que construir **una clase Tablero**. Para facilitar el desarrollo, la mejor opcin es desarrollar una clase tablero donde implementes las siguientes funcionalidades:
    * Cuando se inicialice debers asignar
        * Un id de jugador, para saber de quin es el tablero.
        * Unas dimensiones de tablero, que en el fondo sern tus constantes 10 x 10.
        * Unos barcos. Los que hayas definido como constantes. Aqui simplemente puedes pasar, por ejemplo, un diccionario donde especifiques el nombre de tus barcos, y la eslora de cada uno. Luego ya los colocars en el tablero.
        * **Un tablero sin barcos, que ser un array de `numpy`** donde posicionars los barcos. Este tablero est vaco, por lo que lo puedes rellenar de 0s, 1s, o el caracter que consideres.
        * Adicionalmente la clase tablero necesitar otro array de `numpy`, por qu? porque el tablero de la maquina tendr internamente un array con sus barcos (lo que no vemos) y har falta otro array (que s veremos nosotros) con los disparos efectuados, para saber dnde tenemos que disparar.
    * **Inicializar el tablero**, es decir, colocar los barcos. Puedes pasar por alto el hecho de que tengan que tener espacios entre ellos pero si los colocas aleatoriamente, mucho cuidado aqu de poner los barcos dentro del tablero, y de no colocar unos barcos encima de otros :)
    * Necesitars un mtodo de **disparo coordenada**. Cuando hay un disparo de un jugador en ese tablero, tendrs que comprobar si ahi haba un barco, o simplemente agua. Acurdate de marcar en el tablero, tanto si hay un impacto, como si dio agua.
    * NO te cias a los mtodos que te acabo de mencionar, crea todos los que necesites, introduce en el constructor lo que quieras y desarrolla las funciones que consideres oportunas para facilitarte el desarrollo.

3. Una vez ya tienes modelizado tu tablero, hay que montar el programa que se ejecutara desde un **main.py**:
    * El programa no es ms que el **tpico `while true: `, con una serie de inputs del usuario**. Se est ejecutando constantemente y le pide al usuario coordenadas para comprobar si impacta.
    * Cuando arranque el programa, primero pon algn mensaje de bienvenida y las instrucciones del juego.
    * A continuacin **inicializa los tableros de ambos jugadores** con los barcos. Estas dos primeras acciones solo se ejecutan una vez!! Que es el comienzo del juego.
    * Despus de eso ya comienza el juego. Bsicamente **se ir ejecutando iterativamente en el `while`, y le ir preguntando coordenadas al usuario.**
    * Recoges coordenadas, compruebas en el tablero de la mquina si habia barco.
        * Hay barco: marca en el tablero de la maquina el impacto y le vuelve a tocar al usuario
        * No hay barco: le toca a la maquina. O lo que es lo mismo, escoge una coordenada aleatoria, y comprueba en el tablero del usuario si habia barco.
    * **As hasta que uno de los dos jugadores se quede sin barcos, y termina el juego.**
    * Cuando empiece tu turno deberas imprimir por pantalla tu tablero, para ver cuntos impactos te ha hecho la mquina, as como el tablero con los impactos que has hecho tu en el adversario, de manera que te sirva de ayuda para el siguiente disparo.
    * Todas aquellas funciones que puedas construir para la ejecucin de este programa debern estar definidas en un script que se llame **funciones.py**.

Hay infinidad de maneras para resolver este ejercicio, sentos libres de implementarlo de la forma que te resulte ms cmoda.



## Recomendaciones
1. **No es necesario tener una clase barco, aunque es recomendable** con los barcos de cada jugador, cules estn tocados y dems. Simplemente con llevar un inventario de las coordenadas con barco es suficiente. Por ejemplo, si el tablero tuviese 2 barcos de 4 posiciones de eslora, es como si tuvieses 8 vidas. Cuando se te acaban las vidas, pierdes.
2. Para implementar el bucle `while`, puedes usar **sentencias `continue`, `break`**.
3. Se recomienda tener el cdigo separado en **varios scripts**. Por ejemplo:
    * main.py: donde corre todo el programa
    * clases.py: aqu irn las clases
    * funciones.py: aqu irn las funciones
    * variables.py: donde estn declaradas las constantes
    * A gusto del desarrollador...
4. Cuando se imprime el tablero por pantalla, bsicamente ser un `print()` del array de `numpy`. Escoge con sentido qu quieres que se imprima por pantalla, de manera que se vea bien dnde hay barco, donde ha habido impacto, agua... Tendrs los siguients casos:
    * Agua donde no se ha impactado
    * Agua donde se ha impactado
    * Barco donde no se ha impactado
    * Barco donde se ha impactado
5. Utiliza el `debugging` del IDE de Visual Studio Code si ves que tu programa no se comporta como debera y no sabes muy bien por qu.
6. Se recomienda en la clase tablero manejar dos arrays de 10x10. Uno con las coordenadas de los barcos y los impactos, que ser tu tablero, lo que veas tu cuando imprimes tu tablero por pantalla. Y otro tablero SIN los barcos, nicamente con los impactos, para ver qu impactos has hecho en el tablero de la maquina.
7. Si vas a posicionar los barcos de manera aleatoria, tendrs que escoger unas coordenadas aleatoriamente, y una orientacin N/S/E/O, de manera que si tu barco tiene 3 de eslora y la orientacin es N, se posicionar hacia arriba, desde la posicin inicial elegida aleatoriamente.
8. Si no tienes un nivel alto de Python, prueba primero a crear el array de numpy vaco y a colocar los barcos en unas posiciones concretas que nunca cambien, porque si lo haces aleatoriamente te ser ms complicado. No es justo como funciona el juego, pero mejor avanzar a quedarnos atascados al principio.
