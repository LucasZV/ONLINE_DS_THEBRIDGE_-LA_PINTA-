## PRACTICA GRUPAL FINAL: Despliegue Modelo

### Objetivo

Desplegar un modelo accesible mediante API Rest en algn entorno accesible a travs de una direccin pblica de internet (se sugiere PythonAnywhere por ser el visto en clase, pero es completamente voluntario). Por accesible se entiende poder hacer una peticin con datos de entrada que permitan dar una prediccin o etiquetado al modelo y que el servicio devuelva la etiqueta o prediccin generada por el modelo. Esta funcionalidad se debe ofrecer a travs de un endpoint accesible al menos mediante una peticin requests.get de Python. 

Como completmento al anterior endpoint se debe proveer un endopoint ligado al "/" o landingpage que informe de cmo se deben acceder el resto de endpoints (si hay alguno ms adems del que permita el uso del modelo).

Finalmente, durante la exposicin del trabajo adems de hacer una prueba in-situ del uso de los endpoints proporcionados por la web-app del grupo, se pedir hacer modificacin en el cdigo y hacer un redespliegue de forma que se implemente dicho cambio (generalmente un cambio en un texto, algo sencillo)

El redespliegue puede ser manual. Pero tanto el primer despliegue como el redespliegue han de ser sin tocar ms cdigo en el servidor que el de configuracin del servidor web. 

### Extra voluntario

Se pide como extra voluntario: 
- Implementar un mecanismos de despliegue continuo mediante webhooks. Es decir que el despliegue se haga de forma continua y sin tocar en el servidor una vez "comprometidos" (commited) los cambios en el repositorio del grupo.
- Implementar un endpoint que permitar el reentrenamiento del modelo, aadiendo un dataset nuevo de entrenamiento al repositorio del cdigo de la app.


### Detalles adicionales

Observa que detrs de tanto texto no hay ms ni menos que repetir los tutoriales y talleres de las sesiones en vivo de despliegue de APIS. La recomendacin principal es que desde el principio utilicis un repositorio como forma de subir la informacin al sevidor mediante clone y pulls sucesivos ya que como bien recuerdas "Nunca tocar el cdigo en produccin"... Esto no incluye los ficheros de configuracin necesarios en el servidor.
