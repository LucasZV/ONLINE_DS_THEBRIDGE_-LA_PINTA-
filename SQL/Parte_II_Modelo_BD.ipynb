
# Prctica Grupal: Diseo de Base de Datos

Para esta prctica se pide disear el modelo de datos con las tablas SQL, sus campos y tipos as como las relaciones de una base datos cuyas caractersticas se detallan a continuacin.

## Especificaciones

Tenemos que disear una base de datos sobre proveedores y disponemos de la siguiente
informacin.

De cada proveedor conocemos:  
 Nombre  
 Direccin  
 Ciudad  
 Provincia  
 Cdigo de proveedor (nico para cada uno de ellos)  
  
Queremos llevar un control de las piezas que nos suministra cada proveedor. Es importante
conocer la cantidad de las diferentes piezas que nos suministra y en qu fecha lo hace. Un
mismo proveedor nos puede suministrar una pieza con el mismo cdigo en diferentes
fechas.  

El diseo de la base de datos debe permitir almacenar un histrico con todas las fechas y
las cantidades que nos ha proporcionado un proveedor. Una misma pieza puede ser
suministrada por diferentes proveedores.  

De cada pieza conocemos:
- Cdigo de pieza(nico)
- Nombre
- Color
- Precio
- Categora.  

Pueden existir varias categoras y cada categora tiene:
- Nombre
- Cdigo de categora nico
Una pieza slo puede pertenecer a una categora.
